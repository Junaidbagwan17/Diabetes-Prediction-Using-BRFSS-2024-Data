###  Cleaning Invalid Codes and Handling Missing Values  

Several BRFSS variables contain special numeric codes such as **77, 88, and 99**, which represent:
- *77 = Don't know / Not sure*  
- *88 = None / No days*  
- *99 = Refused / Missing information*  

The important thing is these codes are not actual data values and must be treated as missing values.

####  Steps Performed:
1. **Checked unique values** for each variable to identify the presence of invalid codes.  
2. **Replaced invalid codes (77, 88, 99)** with `NaN` to ensure consistency in missing-value handling.  
3. **Filled missing values using appropriate imputation methods** such as:
   - **Mode** for categorical variables, there are more binary values so mode is suitable
   - **Median** for numerical variables    
4. **Re-checked updated values** to confirm that invalid codes were removed and replaced correctly.

This approach prevents unnecessary loss of data, especially since dropping all rows with NaN would significantly reduce the dataset size. Imputing missing values maintains dataset integrity and supports better model performance.


```python
# 1 Check unique values (including NaN)
print(df['GENHLTH'].unique())

# 2 Replace invalid codes with NaN
df['GENHLTH'] = df['GENHLTH'].replace([7, 9], np.nan)


# 3 Fill NaN with mode (most frequent value)
df['GENHLTH'] = df['GENHLTH'].fillna(df['GENHLTH'].mode()[0])

#  Check updated values
print(df['GENHLTH'].value_counts(dropna=False))
print(df['GENHLTH'].unique())
```
