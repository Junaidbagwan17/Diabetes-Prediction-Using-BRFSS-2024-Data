    In our data BRFSS, _AGEG5YR is an age category, not the actual age.
    to convert age categories into representative numeric AGE we use (midpoints) of each category.
    eg. Category 1 (0–4 years) -- mid age = 2
    If you leave 1, 2, 3, 4… as raw numbers, the model incorrectly assumes:
    Category 4 > Category 3 (true)
    Midpoints represent approximate real age, This gives models a more realistic representation of age, improves interpretability, and avoids misleading ordinal relationships between category codes.

```python
age_mapping = {
    1: 2,    # 0-4 → 2
    2: 7,    # 5-9 → 7
    3: 12,   # 10-14 → 12
    4: 17,   # 15-19 → 17
    5: 22,   # 20-24 → 22
    6: 27,   # 25-29 → 27
    7: 32,   # 30-34 → 32
    8: 37,   # 35-39 → 37
    9: 42,   # 40-44 → 42
    10: 47,  # 45-49 → 47
    11: 52,  # 50-54 → 52
    12: 57,  # 55-59 → 57
    13: 62   # 60-64 → 62
}
df['_AGEG5YR'] = df['_AGEG5YR'].map(age_mapping)
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols.remove('DIABETE4')
```

    DIABETE4 is my target variable, not a feature.
    I removed it to ensure it does not get scaled, transformed, or used in feature engineering steps that apply only to input features.

## dataset has numeric columns like:
        Age
        BMI
        Mental health days
        Physical health days
### if you don’t scale them:
        Features with larger ranges dominate
        Gradient-based models behave poorly
        Logistic regression coefficients become unstable
        KNN and SVM perform wrongly
        Models converge slowly

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

    Scaled data Because my numeric features are on very different scales.
    I used StandardScaler to standardize them to mean = 0 and standard deviation = 1.
    This improves model performance, prevents domination by larger-scale features, and ensures faster and more stable training.
